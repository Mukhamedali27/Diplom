```python
df_2_csv = pd.read_csv('/content/df_2_csv.csv')
df_2_csv.head()
```

```python
columns_to_extract = ['Unnamed: 10', 'Unnamed: 11', 'Unnamed: 15', '0']
selected_columns_df_2 = df_2_csv[columns_to_extract]
selected_columns_df_2.head()
```

```python
selected_columns_df_2.info()
```

```python
pattern = re.compile(r'Договор \(прямой\) (\d{12})')

# Function to extract unique ID from row
def extract_id(row):
    ids = set()
    for col in ['Unnamed: 10', 'Unnamed: 11', 'Unnamed: 15']:
        cell = str(row[col])  # Ensure the cell content is a string
        match = pattern.search(cell)
        if match:
            ids.add(match.group(1))
    return ids.pop() if ids else None

results = []
for _, row in selected_columns_df_2.iterrows():
    id_ = extract_id(row)
    amount = row['0']
    if id_:
        results.append({'ID': id_, 'Amount': amount})


results_df_2 = pd.DataFrame(results)

print(results_df_2)
```

```python
results_df_2.info()
```

```python
results_df_2 = results_df_2.groupby('ID', as_index=False).sum()
results_df_2.info()
```

```python
results_df_2.head()
```
